```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,500)
signal = np.sin(2*np.pi*10*t)
plt.plot(t, signal)
plt.title('Alpha Waves / Stage 1 NREM Sleep Waves')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0,1,500)
signal1 = np.sin(2*np.pi*10*t)
signal2 = np.sin(2*np.pi*5*t) * 0.5
plt.plot(t, signal1+signal2)
plt.title('Alpha and Theta Waves')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
mean = 0
std = 1
N = 1000
noise = np.random.normal(mean, std, N)
fs = 1000  # Sampling frequency
t = np.linspace(0,1,100)
signal1 = np.sin(2*np.pi*10*t)
signal2 = np.sin(2*np.pi*6*t)
noisy_signal = signal1 + signal2 + noise[:100]
plt.plot(t, noisy_signal)
plt.title('Alpha and Theta Waves with Noisy Signal (100 Hz)')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# FFT
N = len(noisy_signal)
fft_vals = np.fft.fft(noisy_signal)
fft_freq = np.fft.fftfreq(N, d=1/fs)

# Use only positive frequencies
positive_freqs = fft_freq[:N//2]
magnitude = np.abs(fft_vals[:N//2])

# Plot
plt.figure()
plt.plot(positive_freqs, magnitude)
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("FFT of Noisy Alpha + Theta Signal (100 Hz Noise)")
plt.xlim(0, 20)   # focus on brain-wave range
plt.show()
```
